The <strong>Customer Churn Prediction Dataset</strong> is a synthetically generated collection that simulates real-world telecom customer data. It includes thousands of customer records, each described by <strong>nine key features</strong> used to predict whether a customer will churn (leave) or stay. The dataset is ideal for building predictive models and understanding customer retention patterns in the telecom industry.

1. üîπ <strong>Key Features:</strong>
    1. <strong>CustomerID</strong>: Unique identifier for each customer.
    2. <strong>Gender</strong>: Customer's gender ‚Äì Male or Female.
    3. <strong>SeniorCitizen</strong>: Indicates whether the customer is a senior citizen (1) or not (0).
    4. <strong>Tenure</strong>: Number of months the customer has been with the service.
    5. <strong>MonthlyCharges</strong>: Monthly billing amount for the customer.
    6. <strong>TotalCharges</strong>: Total amount billed (calculated as Tenure √ó MonthlyCharges).
    7. <strong>Contract</strong>: Type of contract ‚Äì Month-to-month, One year, or Two year.
    8. <strong>PaymentMethod</strong>: Customer‚Äôs payment method ‚Äì Electronic check, Mailed check, Bank transfer, or Credit card.
    9. <strong>Churn</strong>: Indicates if the customer has churned (1) or not (0). This is the <strong>target variable</strong>.

2. üìÅ <strong>Dataset Summary:</strong>
    - <strong>Source</strong>: Synthetic dataset generated using NumPy and Pandas in Python.

## 1. Synthetic Dataset Creation for Customer Churn Prediction

```python
np.random.seed(42)

n = 10000

data = pd.DataFrame({

  'CustomerID': np.arange(n),

  'Gender': np.random.choice(['Male', 'Female'], size=n),

  'SeniorCitizen': np.random.choice([0, 1], size=n),

  'Tenure': np.random.randint(1, 72, size=n),

  'MonthlyCharges': np.round(np.random.uniform(20, 120, size=n), 2),

  'TotalCharges': lambda df: df['Tenure'] * df['MonthlyCharges'],

  'Contract': np.random.choice(['Month-to-month', 'One year', 'Two year'], size=n),

  'PaymentMethod': np.random.choice(['Electronic check', 'Mailed check', 'Bank transfer', 'Credit card'], size=n),

  'Churn': np.random.choice([0, 1], size=n, p=[0.73, 0.27])

})

data['TotalCharges'] = (data['Tenure'] * data['MonthlyCharges']).round(2)

data.to_csv('data/customer_churn_data.csv', index=False)
```

## 1. Univariate analysis for categorical variables

```python
# Set Seaborn style
sns.set(style='whitegrid')
palette = sns.color_palette("Set2")

# Loop through categorical columns
for col in categorical_columns:
    plt.figure(figsize=(8, 5))
    ax = sns.countplot(data=df, x=col, palette=palette)
    
    total = len(df)
    for p in ax.patches:
        count = p.get_height()
        percentage = f'{100 * count / total:.1f}%'
        x = p.get_x() + p.get_width() / 2
        y = p.get_height() / 2  # Mid-bar position
        ax.annotate(percentage, (x, y), ha='center', va='center', fontsize=11, color='black', fontweight='bold')
    
    plt.title(f'Distribution of {col}', fontsize=14, fontweight='bold', color='darkblue')
    plt.xlabel(col, fontsize=12)
    plt.ylabel('Count', fontsize=12)
    plt.xticks(rotation=45)

    # Add legend for categories
    handles, labels = ax.get_legend_handles_labels()
    if not handles:  # If legend not automatically added (e.g., for Gender), add manually
        unique_vals = df[col].unique()
        custom_handles = [plt.Rectangle((0,0),1,1, color=palette[i]) for i in range(len(unique_vals))]
        plt.legend(custom_handles, unique_vals, title=col, bbox_to_anchor=(1.05, 1), loc='upper left')
    
    plt.tight_layout()
    plt.show()
```

### üî∑ 1. Correlation Heatmap of Numerical Features

#### üîç Insights:
- **Tenure and TotalCharges** have a **strong positive correlation** (`0.77`). Customers with longer tenure tend to have higher total charges.

### ‚úÖ Summary
