<h2 style="color:#D00000; margin-bottom: 20px;">‚úÖ Missing Values Analysis</h2>

After carefully reviewing the dataset, the following decisions were made regarding missing and unnecessary data:

1. **Missing Values in `TotalCharges`:**
   - There were **11 missing values** detected in the `TotalCharges` column.
   - Upon investigation, it was found that in **all 11 cases**, the value of `Tenure` was **0**, which indicates that these customers had just joined and had not been billed yet.
   - Filling these rows using `Tenure * MonthlyCharges` would result in a `TotalCharges` value of **0**, which adds no meaningful insight.
   - ‚úÖ **Action Taken:** All 11 rows with missing `TotalCharges` were **dropped** to ensure data quality and modeling reliability.

2. **Dropping Irrelevant Column - `CustomerID`:**
   - The `CustomerID` column was a unique identifier and carried **no predictive value** for churn modeling.
   - ‚úÖ **Action Taken:** The `CustomerID` column was **removed** from the dataset to reduce dimensionality and simplify preprocessing.

‚úÖ Final Integrity Check:
- No missing values remain in the dataset.
- All columns retained are relevant for customer churn prediction.

> üí° These steps help improve model focus and avoid bias or noise from irrelevant or incomplete records.

## Getting X and y variables

## Creating Pipeline with Column Transformer

```python
categorical_cols = X.select_dtypes(include='object').columns
numerical_cols = X.select_dtypes(exclude='object').columns
```

```python
categorical_cols
```

```python
numerical_cols
```

```python
# Numerical Pipeline
num_pipeline = Pipeline(
                steps = [
                ('imputer',SimpleImputer(strategy='median')),
                ('scaler',StandardScaler())                
                ]
            )
num_pipeline
```

```python
# Categorical Pipeline
cat_pipeline = Pipeline(
                steps=[
                ('imputer',SimpleImputer(strategy='most_frequent')),
                ('label_encoder',OrdinalEncoder()),
                # ('scaler',StandardScaler())
                ]
            )

cat_pipeline
```

```python
preprocessor = ColumnTransformer(
                [
                ('num_pipeline',num_pipeline,numerical_cols),
                ('cat_pipeline',cat_pipeline,categorical_cols)
                ]
            )
preprocessor
```

## Train Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

## Training Various models

```python
# Define models
models = {
    'LogisticRegression': LogisticRegression(),
    'SVC': SVC(),
    'KNN': KNeighborsClassifier(),
    'NaiveBayes': GaussianNB(),
    'DecisionTree': DecisionTreeClassifier(),
    'RandomForest': RandomForestClassifier(),
    'AdaBoost': AdaBoostClassifier(),
    'GradientBoosting': GradientBoostingClassifier(),
    'XGBoost': XGBClassifier(),
    'CatBoost': CatBoostClassifier(verbose=False),
    'LightGBM': LGBMClassifier()
}
models
```

```python
model_list = []
accuracy_scores = []

for name, model in models.items():
    model.fit(X_train, y_train)

    # Predictions
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)

    # Metrics
    train_acc = accuracy_score(y_train, y_train_pred)
    test_acc = accuracy_score(y_test, y_test_pred)

    train_cm = confusion_matrix(y_train, y_train_pred)
    test_cm = confusion_matrix(y_test, y_test_pred)

    train_cr = classification_report(y_train, y_train_pred, output_dict=True)
    test_cr = classification_report(y_test, y_test_pred, output_dict=True)

    # Save model name and accuracy
    model_list.append(name)
    accuracy_scores.append(test_acc)

    print(f"\nüîç Model: {name}")
    print(f"‚úîÔ∏è Train Accuracy: {train_acc:.4f}")
    print(f"‚úîÔ∏è Test Accuracy : {test_acc:.4f}")

    # Confusion Matrices
    fig, ax = plt.subplots(1, 2, figsize=(14, 5))
    sns.heatmap(train_cm, annot=True, fmt='d', cmap='Blues', ax=ax[0])
    ax[0].set_title(f'{name} - Train Confusion Matrix')
    ax[0].set_xlabel('Predicted')
    ax[0].set_ylabel('Actual')

    sns.heatmap(test_cm, annot=True, fmt='d', cmap='Greens', ax=ax[1])
    ax[1].set_title(f'{name} - Test Confusion Matrix')
    ax[1].set_xlabel('Predicted')
    ax[1].set_ylabel('Actual')

    plt.tight_layout()
    plt.show()

    # Classification Report (Test Set)
    cr_df = pd.DataFrame(test_cr).transpose()
    display(cr_df.style.background_gradient(cmap='YlGnBu').format("{:.2f}"))

    # ROC Curve (if model supports probabilities)
    if hasattr(model, "predict_proba"):
        y_proba = model.predict_proba(X_test)[:, 1]
        fpr, tpr, _ = roc_curve(y_test, y_proba)
        roc_auc = auc(fpr, tpr)

        plt.figure(figsize=(8, 6))
        plt.plot(fpr, tpr, color='darkorange', lw=2, label=f"ROC Curve (AUC = {roc_auc:.2f})")
        plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
        plt.xlabel("False Positive Rate")
        plt.ylabel("True Positive Rate")
        plt.title(f"ROC Curve - {name}")
        plt.legend(loc="lower right")
        plt.grid(True)
        plt.tight_layout()
        plt.show()
    else:
        print(f"‚ö†Ô∏è ROC Curve not supported for {name} (no predict_proba method)")

    print("=" * 100)
```

```python
# Best models container
best_estimators = {}
model_scores = {}

for name, (model, params) in models_params.items():
    print(f"\nüîß Tuning {name}...")
    grid = GridSearchCV(model, params, cv=5, scoring='accuracy', n_jobs=-1)
    grid.fit(X_train, y_train)

    best_estimators[name] = grid.best_estimator_
    model_scores[name] = grid.best_score_

    print(f"‚úÖ Best Estimator for {name}:")
    print(grid.best_estimator_)
    print(f"üìà Best Params: {grid.best_params_}")
    print(f"‚≠ê Best CV Score: {grid.best_score_:.4f}")
```

```python
# Find best model based on CV accuracy
best_model_name = max(model_scores, key=model_scores.get)
print(f"\nüèÜ Best Model Overall: {best_model_name}")
print(f"üìä Accuracy: {model_scores[best_model_name]:.4f}")
print(f"üîç Estimator: {best_estimators[best_model_name]}")
```
